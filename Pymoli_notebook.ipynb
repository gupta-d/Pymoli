### Purchasing Analysis by Gender

```python
# a few of below commands return pd.series data object which can be accessed like a list. 
#To get a specific value use [<integer index of the list item >] / or[<name index of the list item>]. using names

total_purchase_count_male = df.loc[df['Gender'] == 'Male', :].count()['SN']
average_purchase_price_male =  df.loc[df.loc[:,'Gender'] =='Male', ['Price']].mean()['Price']
total_purchase_value_male = df.loc[df.loc[:,'Gender'] =='Male', ['Price']].sum()['Price']
number_of_unique_male_players = df.loc[df.loc[:,'Gender'] =='Male', :].nunique()['SN']

total_purchase_count_female = df.loc[df['Gender'] == 'Female', :].count()['SN']
average_purchase_price_female =  df.loc[df.loc[:,'Gender'] =='Female', ['Price']].mean()['Price']
total_purchase_value_female = df.loc[df.loc[:,'Gender'] =='Female', ['Price']].sum()['Price']
number_of_unique_female_players = df.loc[df.loc[:,'Gender'] =='Female', :].nunique()['SN']

total_purchase_count_other = df.loc[df['Gender'] == 'Other / Non-Disclosed', :].count()['SN']
average_purchase_price_other =  df.loc[df.loc[:,'Gender'] =='Other / Non-Disclosed', ['Price']].mean()['Price']
total_purchase_value_other = df.loc[df.loc[:,'Gender'] =='Other / Non-Disclosed', ['Price']].sum()['Price']
number_of_unique_other_players = df.loc[df.loc[:,'Gender'] =='Other / Non-Disclosed', :].nunique()['SN']
```

### Age Demographics
Option1: Without Pivot Table

```python
df['Age'].max()
```

```python
bins = [0,10, 14, 19, 24, 29, 34, 39, 40]
age_labels = ['0-10','10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
df['Age Bracket'] = pd.cut (df['Age'], bins, labels= age_labels) # assignment adds new column, hence not needing 'inplace'
```

```python
# pandas accepts  dictionary as 1 argument while creating a dataframe
#creating a dictionary for age demographics (pertaining to columns of dataframe)
# index will be passed on separately as age-labels while creating dataframe

dict_age_demo = {'Purchase Count':[],'Total Purchase Value':[], 'Average Purchase Price': [], 'Average Purchase Total per Person by Age Group': []  } 

for label in age_labels:
    count = (df.loc[df['Age Bracket'] == label, :].count()['Age Bracket'])
    unique_count = (df.loc[df['Age Bracket'] == label, :].nunique()['SN'])
    total = (df.loc[df['Age Bracket'] == label, :].sum()['Price'])
    average_price = total/count
    average_purchase_total_per_person = total/unique_count
    dict_age_demo['Purchase Count'].append (count)
    dict_age_demo['Total Purchase Value'].append(total)
    dict_age_demo['Average Purchase Price'].append(average_price)
    dict_age_demo['Average Purchase Total per Person by Age Group'].append(average_purchase_total_per_person)
# dict1.update((label, (df.loc[df['Age Bracket'] == label, :].count()['Age Bracket']))
```

```python
df_age_demo= pd.DataFrame(dict_age_demo, index = age_labels)
df_age_demo
```

```python
# Formatting output 
df_age_demo['Total Purchase Value'] = df_age_demo['Total Purchase Value'].map('${:,.2f}'.format)
df_age_demo['Average Purchase Price'] = df_age_demo['Average Purchase Price'].map('${:,.2f}'.format)
df_age_demo['Average Purchase Total per Person by Age Group'] = df_age_demo['Average Purchase Total per Person by Age Group'].map('${:,.2f}'.format)
df_age_demo
```

### Age Demographics
Option2: Using Pivot Table (Much Simpler...)

```python
#DataFrame.pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All')[source]Â¶
df_pivot = df.pivot_table( values = ['Price', 'SN'], index='Age Bracket',  aggfunc={'Price':['count','mean', 'sum'], 'SN':'nunique'})
df_pivot
```

```python
# adding a column for Average Purchase Total per Person
df_pivot.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Number of Players']

# updating the new column with calculated average price per person
df_pivot['Average Purchase Total per Person'] = df_pivot['Total Purchase Value']/df_pivot['Number of Players']
```

```python
# formatting the dataframe with $ sign and decimal upto 2 places
df_pivot['Total Purchase Value'] = df_pivot['Total Purchase Value'].map('${:,.2f}'.format)
df_pivot['Average Purchase Price'] = df_pivot['Average Purchase Price'].map('${:,.2f}'.format)
df_pivot['Average Purchase Total per Person'] = df_pivot['Average Purchase Total per Person'].map('${:,.2f}'.format)
df_pivot
```

### Top Spenders

```python
# Groupby player name and sum all columns
df_spenders = df.groupby(by='SN').sum()
df_spenders.head()
# return object of groupby will have 'SN' as indexes

```

```python
# creating 3 dataframe objects from groupby function 

df_spenders = df.groupby(by='SN').sum()
df_popular = df.groupby(by='Item ID').count()
df_profitable = df.groupby(by='Item ID').sum()
```

```python
# Option 1: sorting dataframe objects to get top-5 rows on top
df_spenders = df_spenders.sort_values(by= 'Price', ascending = False)
df_popular = df_popular.sort_values(by= 'Item Name', ascending = False)
df_profitable = df_profitable.sort_values(by= 'Price', ascending = False)

# print top-5 rows to verify
df_spenders.head(5)
df_popular.head(5)
df_profitable.head(5)
```

Now we will prepare the results in dictionaries and convert to dataframes in below cells

```python
#creating lists of indexes for top-5 items
top5_spenders_list = [df_spenders.index[i] for i in range (5)]
top5_popular_list = [df_popular.index[i] for i in range (5)]
top5_profitable_list = [df_profitable.index[i] for i in range (5)]

# or rather than sorting the groupby result, we can simply use pandas native method ".largest" as shown below
#top_spenders_list = df_spenders['Price'].nlargest(5).index.tolist()
```

```python
# make dataframe for top5 spenders
dict2 = {'SN':[], 'Purchase Count': [], 'Average Purchase': [], 'Total Purchase Value':[]}
for i in range(5):
    name = df_spenders.index[i]
    count = df.loc[df['SN'] == name, :].count()['Price']
    total_price = df.loc[df['SN'] == name, :].sum()['Price']
    average_price = total_price/count

    
    dict2['SN'].append(name)
    dict2['Purchase Count'].append(count)
    dict2['Average Purchase'].append(average_price)
    dict2['Total Purchase Value'].append(total_price)
    
top5_spenders = pd.DataFrame(dict2)
top5_spenders.set_index('SN', inplace = True)
top5_spenders['Average Purchase'] = top5_spenders['Average Purchase'].map('${:,.2f}'.format)
top5_spenders['Total Purchase Value'] = top5_spenders['Total Purchase Value'].map('${:,.2f}'.format)
top5_spenders
```

```python
# make dataframe for top5 popular items
# creating a dictionary structure with empty lists as present values
dict3 = {'Item ID':[],'Item Name':[], 'Purchase Count': [], 'Item Price': [], 'Total Purchase Value':[]}

def item_name(ID): # function to return Item name against an item ID . df.at or df.loc both can be used
    for i in range (len(df)):
        if df.at[i, 'Item ID'] == ID:
            return df.at[i, 'Item Name']
            break
        else:
            pass

def item_price(ID): # function to return Item unit price against an item ID 
    for i in range (len(df)):
        if df.at[i, 'Item ID'] == ID:
            return df.at[i, 'Price']
            break
        else:
            pass

for i in range(5): # Getting particulars for top-5 popular items and store in a dictionary
    ID = df_popular.index[i]
    
    # using original dataframe to get required details for all popular items 
    name = item_name(ID)
    count = df.loc[df['Item ID'] == ID, :].count()['Price']
    itemized_price = item_price(ID)
    total_price = df.loc[df['Item ID'] == ID, :].sum()['Price']
        
    dict3['Item ID'].append(ID)
    dict3['Item Name'].append(name)
    dict3['Purchase Count'].append(count)
    dict3['Item Price'].append(itemized_price)
    dict3['Total Purchase Value'].append(total_price)
    
    
top5_popular = pd.DataFrame(dict3)
top5_popular.set_index(['Item ID' , 'Item Name'], inplace = True)
top5_popular['Item Price'] = top5_popular['Item Price'].map('${:,.2f}'.format)
top5_popular['Total Purchase Value'] = top5_popular['Total Purchase Value'].map('${:,.2f}'.format)

top5_popular
```

```python
# make dataframe for top5 profitable items
# creating a dictionary structure 
dict4 = {'Item ID':[],'Item Name':[], 'Purchase Count': [], 'Item Price': [], 'Total Purchase Value':[]}

for i in range(5): # Getting particulars for top-5 popular items and storint in dictionary
    ID = df_profitable.index[i]
    name = item_name(ID)# function already defined above
    count = df.loc[df['Item ID'] == ID, :].count()['Price']
    itemized_price = item_price(ID) # functional already defined above
    total_price = df.loc[df['Item ID'] == ID, :].sum()['Price']
        
    dict4['Item ID'].append(ID)
    dict4['Item Name'].append(name)
    dict4['Purchase Count'].append(count)
    dict4['Item Price'].append(itemized_price)
    dict4['Total Purchase Value'].append(total_price)
    
    
top5_profitable = pd.DataFrame(dict4)
top5_profitable.set_index(['Item ID' , 'Item Name'], inplace = True)
top5_profitable['Item Price'] = top5_profitable['Item Price'].map('${:,.2f}'.format)
top5_profitable['Total Purchase Value'] = top5_profitable['Total Purchase Value'].map('${:,.2f}'.format)

top5_profitable
```
